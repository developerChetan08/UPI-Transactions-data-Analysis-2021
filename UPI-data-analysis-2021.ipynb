```python
data = pd.read_csv('UPI apps transaction data in 2021.csv')
```

# Top 10 highest volume payment by UPI Banks

Volume (Mn) By Costumers

```python
total = data.groupby(['UPI Banks']).agg({'Volume (Mn) By Costumers':'sum'}).reset_index()
```

```python
top_10_highest_volume_mn = total.sort_values(by = 'Volume (Mn) By Costumers',ascending = False).head(10)
top_10_highest_volume_mn
```

Value (Cr) by Costumers

```python
total = data.groupby(['UPI Banks']).agg({'Value (Cr) by Costumers':'sum'}).reset_index()
top_10_highest_volume_cr = total.sort_values(by = 'Value (Cr) by Costumers',ascending = False).head(10)
top_10_highest_volume_cr
```

Volume (Mn)

```python
total = data.groupby(['UPI Banks']).agg({'Volume (Mn)':'sum'}).reset_index()
top_10_highest_volume_Mn = total.sort_values(by = 'Volume (Mn)',ascending = False).head(10)
top_10_highest_volume_Mn
```

Value (Cr)

```python
total = data.groupby(['UPI Banks']).agg({'Value (Cr)':'sum'}).reset_index()
top_10_highest_volume_Cr = total.sort_values(by = 'Value (Cr)',ascending = False).head(10)
top_10_highest_volume_Cr
```
